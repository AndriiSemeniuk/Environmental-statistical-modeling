```python
from scipy.stats import norm, binom
```

```python
# parameters of the normal distribution
mu = 10.5   # mean
sigma = 2   # standard deviation
```

```python
# probability that X > 11
p = 1 - norm.cdf(11, loc=mu, scale=sigma)
print("P(X > 11) =", round(p, 4))
```

```python
# binomial probability: exactly 2 successes out of 4
n = 4
k = 2
prob = binom.pmf(k, n, p)
print("P(exactly 2 out of 4 years > 11) =", round(prob, 4))
```
