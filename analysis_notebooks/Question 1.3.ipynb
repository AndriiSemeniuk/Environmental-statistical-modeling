```python
import pandas as pd
```

```python
data=pd.read_csv('Question 3.csv')
```

```python
data.head
```

```python
column = 'Annual freshwater withdrawals, total (billion cubic meters)'
```

```python
#Calculating the mean:
mean_val = data[column].mean()
```

```python
#Calculating the median:
median_val = data[column].median()
```

```python
#Calculating the mode:
mode_val = data[column].mode()[0]
```

```python
#Calculating the variance:
variance_val = data[column].var()
```

```python
#Calculating the Standard Deviation:
std_val = data[column].std()
```

```python
#Calculating the range:
range_val = data[column].max() - data[column].min()
```

```python
#Calculating the IQR (Interquartile Range):
iqr_val = data[column].quantile(0.75) - data[column].quantile(0.25)
```

```python
print(f"Mean: {mean_val}")
print(f"Median: {median_val}")
print(f"Mode: {mode_val}")
print(f"Variance: {variance_val}")
print(f"Standard Deviation: {std_val}")
print(f"Range: {range_val}")
print(f"IQR (Interquartile Range): {iqr_val}")
```
